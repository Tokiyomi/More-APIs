```python
import requests
import json
import pandas as pd
```

```python
def BuscarEntidad(condicion, entidad, registroInicial, registroFinal, token):
    url= "https://www.inegi.org.mx/app/api/denue/v1/consulta/BuscarEntidad/{}/{}/{}/{}/{}".format(condicion, entidad, registroInicial, registroFinal, token)
    res = requests.get(url)
    results = json.loads(res.content)
    return results
```
