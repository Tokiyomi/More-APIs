a. Create the function named *astro_now*() to obtain data of the astraonauts in the space using the method [Number of People in Space](http://open-notify.org/Open-Notify-API/People-In-Space) using the following python libraries:
* [requests](https://2.python-requests.org/en/master/)
* [json](https://docs.python.org/3/library/json.html)
    
The function must return the obtained json file as a python dictionary.

```python
def astro_now():
    """
    Function that makes an API request to get a json file from 'http://api.open-notify.org/astros.json' 
    ------------------
    Input: <none>
    -------------------
    Output: <dict>
        Returns a dictionary including the current number of people in space. 
        When known it also returns the names and spacecraft those people are on.
    """
    endpoint_url='http://api.open-notify.org/astros.json'

    res = requests.get(endpoint_url) #Get the information
    
    res.content #Read it as an object
    
    results=json.loads(res.content) #Convert the object into a dict
    
    return results #Return a dictionary

#API request
astronauts_json = astro_now()

type(astronauts_json)
```

c. Create the function named *iss_currentlocation*() to obtain the position of the International Space Station using the method [Current Location of the International Space Station](http://open-notify.org/Open-Notify-API/ISS-Location-Now/).
    
The function must return the numeric value of the latitud and longitud as a tuple, e.g. (20.0,120.0)

```python
# Complete task c here
def iss_currentlocation():
    """
    Function that obtains the position of the International Space Station 
    using the an API request from a json file 'http://api.open-notify.org/iss-now.json'
    ---------------------
    Input: <none>
    -------------------------------
    Output: <float tuple>
        Returns the numeric value of the latitud and longitud as a tuple.
    """
    endpoint_url='http://api.open-notify.org/iss-now.json'

    res = requests.get(endpoint_url) #Get the information
    
    res.content #Get the file as object
    
    results=json.loads(res.content) #Load the json file as dict
    
    results=results['iss_position'] #Take the key were lat and lon are
    
    lon=float(results['longitude']) #Assign longitude to a float variable
    
    lat=float(results['latitude'])  #Assign latitudee to a float variable
    
    return lon, lat #As tuple

#API request
location_json = iss_currentlocation()
location_json #print tuple
```

d. Use your *iss_currentlocation()* function to obtain the position of the International Space Station each 10 seconds during 5 minutes. Show the results in a table with the following columns:
* timestamp (Parsed Correctly)
* latitude
* longitude

The [Time](https://docs.python.org/3/library/time.html) Python library could be useful.

```python
# Complete task d here
def iss_timestamp():
    """Get the timestamp data from the API url 'http://api.open-notify.org/iss-now.json
    This functions returns a parsed timestamp data from that request
    ---------------------------
    Input:
        <None>
    ---------------------------
    Output: 
        <int>
        Timestamp data (unix).
    """
    endpoint_url = 'http://api.open-notify.org/iss-now.json' #URL of the json file
    
    res = requests.get(endpoint_url) #Get the file (object)
    
    results = json.loads(res.content) #Load the object as a dict
    
    timestamp = results['timestamp'] #select timestamp info
    
    return timestamp

timestamp=iss_timestamp()
timestamp
```

```python
#Function that generates the table
def iss_location_10():
    """
    This function makes a dataframe using the information returned from the two functions:
    iss_timestamp() and iss_currentlocation() every 10 seconds for 5 minutes long to get lat, lon and timestamp
    --------------------------
    Input: 
        <None>
    --------------------------
    Output: 
        <DataFrame>
        A data frame with the lat, long, and timestamp for each request.
    """
    import time # import an useful library for manipulating time (seconds)
    
    data = [] # open an empty list
    
    seconds = 0 # seconds counter
    
    while seconds < 300: # For 5 minutes
        
        location = iss_currentlocation() # Getting the (lon, lat) tuple by calling the function 'iss_currentlocation()'
        
        timestamp = iss_timestamp() # Getting the timestamp  by calling the function 'iss_timestamp()' --> (lon, lat)
        
        data.append([timestamp, location[1], location[0]]) # Append the items to the list in this order: [timestamp, lat, lon]
        
        df = pd.DataFrame(data, columns=['timestamp','latitude','longitude']) #creating the data frame
        
        time.sleep(10) #to make it run every 10 sec
        
        seconds += 10 #to make it to the 5 min
        
    df['timestamp'] = pd.to_datetime(df['timestamp'],unit='s') #changing the unix format of the timestamp to a friendly one
    
    return df 

df = iss_location_10() #Calling the function
df
```

e. The International Space Station is an orbital outpost circling high above out heads. Sometimes it’s overhead, but when? It depends on your location. Given a location on Earth (latitude, longitude, and altitude) the [International Space Station Pass Times](http://open-notify.org/Open-Notify-API/ISS-Location-Now/) API compute the next n number of times that the ISS will be overhead.

The function must return the obtained json file as a python dictionary.

```python
def iss_passtimes (latitude, longitude, altitude = None, number = None):
    """
    This function makes an API request to compute the next n number of times that the 
    ISS will be overhead, and transform it to a dictionary.
    ---------------------------------------
    Input: 
        <float>
        
        latitude: of the location desired.
        
        longitude: of the location desired.
        
        altitude: Not necessary
        
        <int>
        
        number: Not necessary
    --------------------------------------
    Output: 
        <dict>
        A dictionary with the information.
    """
    
    endpoint_url = 'http://api.open-notify.org/iss-pass.json' # URL of the json file
    
    parameters = {
        'lat' : latitude, 
        'lon' : longitude,
        'alt' : altitude, 
        'n' : number
    }
    
    res = requests.get(endpoint_url, params = parameters) #Get the data aplying the designed parameters
    
    results = json.loads(res.content) # Load the object as a dict
    
    return results

iss_merida = iss_passtimes(20.967780, -89.624250, altitude = None, number = None) #Evaluate for this coordinate
iss_merida
```
