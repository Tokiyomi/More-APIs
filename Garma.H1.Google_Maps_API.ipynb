B. Create a function to Find the Place Id given Place Name (String) and Location (Lat/Lont Tuple).

```python
#Using Text Search Request Method
def find_place_request(place_name, location, key):
    """
    Function that takes a text input, and returns a place.
    ----------------------
    Input:
        <string>
        
        'place_name' : The text input can be any kind of Places data, for example, a name, address, or phone number.
        'key' : your API key
        'location' : in the form 'lat,lon'
    -------------------------
    Output:
        <json dictionary> containing information about a place
    """
    
    if type(location)!=str:
        location=str(location[0])+','+str(location[1])
        
    endpoint_url='https://maps.googleapis.com/maps/api/place/findplacefromtext/json'
    
    # Establish parameters
    
    locationbias='circle:100@'+location
    
    parameters = {
        'input' : place_name,
        'inputtype' : 'textquery',
        'key' : key,
        'locationbias' : locationbias,
        'fields' : 'name,place_id'
    }
    
    res = requests.get(endpoint_url, params = parameters) #Get the data aplying the designed parameters
    
    res.content #Read it as an object
    
    results = json.loads(res.content) # Read as dict
    
    print(results)
    
    return results['candidates'][0]['place_id'] # Return the firs candidate (which is the nearest to that coordinate)
```

D. Use the [*Place Details*](https://developers.google.com/places/web-service/details#PlaceDetailsResults) Method to find the Reviews of a Place in English and Spanish. In order to fulfill this item you need to:
* Create a *find_reviews* function with the appropiate parameters. It must return a JSON with the results. 
* Create a table with the following columns: Author Name, Language, Rating, Text, Time.

```python
def find_reviews(place_id, key):
    """
    Function that takes a place_id and returns a json dictionary with that place reviews in english and spanish
    ----------------
    Input:
        <string>
            'place_id' : the place ID
            'key' : your API key
    ----------------
    Output:
        <json dictionary> contaning the place reviews
        
    """
        
    endpoint_url='https://maps.googleapis.com/maps/api/place/details/json'
    
    # Establish parameters
    
    parameters = {
        'key' : key,
        'place_id' : place_id,
        'language' : 'en,es'
    }
    
    res = requests.get(endpoint_url, params = parameters) #Get the data aplying the designed parameters
    
    res.content #Read it as an object
    
    results = json.loads(res.content) # Read as dict
    
    return results # Return the firs candidate (which is the nearest to that coordinate)
```

E. Find similar places to Los Trompos Circuito. Rank them by its "Prominence". In order to fulfill this item you need to:
* Create a *find_similar* function with the appropiate parameters. It must return a JSON with the results.
* Create a table with the following columns: Place Name, Plaece Id, Rating, Total of Use Ratings.

```python
def find_similar(location, radius, place_type, key):
    """
    Function that gets a specific location in a given radius and returns the most prominent places
    ------------------
    Input: <string>
            'location' : in the form 'lat,lon'
            'place_type' : for example 'restaurant'
            'key' : your API key
    -------------------
    Output: 
        <json dictionary> containing a list of the most prominent places and its information
    
    """
    
    endpoint_url='https://maps.googleapis.com/maps/api/place/nearbysearch/json'
    
    # Establish parameters
    
    parameters = {
        'key' : key, 
        'location' : location, 
        'radius' : radius, #Note that radius must not be included if rankby=distance
        'type' : place_type, 
        'rankby' : 'prominence' #Prominence
    }

    res = requests.get(endpoint_url, params = parameters) #Get the data aplying the designed parameters
    
    res.content #Read it as an object
    
    results = json.loads(res.content) # Read as dict
    
    return results #Return json as a dict
```
